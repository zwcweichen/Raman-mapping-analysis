# Raman mapping analysis

This Python-based program, developed using PyQt, is designed to assist in the analysis of Raman mapping measurement data. The application features three main tabs, each serving a distinct purpose in the data analysis process.

The first tab is dedicated to reading and processing the experimental raw data. Here's what you can do:

1. **Data Input**:

    Load the Raman spectra for each spatial point in your mapping.
    Import the corresponding x-axis data for the spectra.


2. **Spectrum Preview**:

    View individual Raman spectra in a preview window.
    Use the zoom in and zoom out buttons to adjust the view for detailed spectrum examination.
    Drag to pan across different parts of the spectrum.


3. **Intensity Summation Mapping**:

    Right-click in the spectrum preview window and select "intensity image threshold".
    Drag across a specific range of the spectrum to define the intensity summation range.
    The program calculates the total intensity within this range for each spatial point.
    This process converts each full Raman spectrum at a given location into a single intensity value, creating a 2D intensity map.
