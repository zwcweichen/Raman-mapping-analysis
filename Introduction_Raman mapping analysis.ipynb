# <font face='Times New Roman'> Raman mapping analysis </font>

<font face = 'Times New Roman'>This Python-based program, developed using PyQt, is designed to assist in the analysis of Raman mapping measurement data. The application features three main tabs, each serving a distinct purpose in the data analysis process.</font>

# <font face = 'Times New Roman'>**Tab1**</font>

<font face = 'Times New Roman'>The first tab is dedicated to reading and processing the experimental raw data. Here's what you can do:</font>

<font face = 'Times New Roman'>1. **Data Input**:</font>

<font face = 'Times New Roman'>    Load the Raman spectra for each spatial point in your mapping.</font>
<font face = 'Times New Roman'>    Import the corresponding x-axis data for the spectra.</font>


<font face = 'Times New Roman'>2. **Spectrum Preview**:</font>

<font face = 'Times New Roman'>    View individual Raman spectra in a preview window.</font>
<font face = 'Times New Roman'>    Use the zoom in and zoom out buttons to adjust the view for detailed spectrum examination.</font>
<font face = 'Times New Roman'>    Drag to pan across different parts of the spectrum.</font>


<font face = 'Times New Roman'>3. **Intensity Summation Mapping**:</font>

<font face = 'Times New Roman'>    Right-click in the spectrum preview window and select "intensity image threshold".</font>
<font face = 'Times New Roman'>    Drag across a specific range of the spectrum to define the intensity summation range.</font>
<font face = 'Times New Roman'>    The program calculates the total intensity within this range for each spatial point.</font>
<font face = 'Times New Roman'>    This process converts each full Raman spectrum at a given location into a single intensity value, creating a 2D intensity map.</font>


<font face = 'Times New Roman'>4. **Interactive Spectrum Exploration**:</font>

<font face = 'Times New Roman'>    After generating the intensity summation map, you can interactively explore individual spectra.</font>
<font face = 'Times New Roman'>    Press and hold the 'A' key on your keyboard.</font>
<font face = 'Times New Roman'>    While holding 'A', move your mouse over the intensity map.</font>
<font face = 'Times New Roman'>    The spectrum corresponding to the current mouse position will be displayed in real-time.</font>
<font face = 'Times New Roman'>    This feature allows for quick and intuitive exploration of spectral data across different spatial locations in your sample.</font>
